```
# --- STEP 1: Import required libraries ---
import pandas as pd
from google.colab import files
```

```
# --- STEP 2: Upload both CSV files manually ---
print("📂 Please upload: metal_prices.csv")
uploaded_financial = files.upload()                    # File picker appears for first file
financial_path = list(uploaded_financial.keys())[0]

print("📂 Please upload: news_impact_score.csv")
uploaded_impact = files.upload()                       # File picker appears for second file
impact_path = list(uploaded_impact.keys())[0]
```

```
# --- STEP 3: Load datasets ---
financial_df = pd.read_csv(financial_path)
impact_df = pd.read_csv(impact_path)
```

```
# --- STEP 4: Data preprocessing ---
# Convert Date columns to datetime
financial_df["Date"] = pd.to_datetime(financial_df["Date"], format="%d-%m-%Y")
impact_df["Date"] = pd.to_datetime(impact_df["Date"])

# Sort both by date
financial_df.sort_values("Date", inplace=True)
impact_df.sort_values("Date", inplace=True)
```

```
# --- STEP 5: Merge by date (outer join ensures all days preserved) ---
merged_df = pd.merge(
    financial_df,
    impact_df,
    on="Date",
    how="left"
)

# Forward-fill and backward-fill missing Impact values (no news on weekends, etc.)
merged_df["Impact"] = merged_df["Impact"].ffill().bfill()
```

```
# --- STEP 6: Save the final merged dataset ---
output_path = "/content/metal_prices_impact.csv"
merged_df.to_csv(output_path, index=False)
print(merged_df.head())

files.download(output_path)   # ← Auto-downloads as metal_prices_impact.csv
print("✅ Download triggered: metal_prices_impact.csv")
```
